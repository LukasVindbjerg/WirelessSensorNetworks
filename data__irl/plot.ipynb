```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = [20, 10]
```

```python
node1_file = "node1.csv"
node2_file = "node2.csv"
node3_file = "node3.csv"
# jammer_file = "random_jam.csv"

node1 = pd.read_csv(node1_file, sep="\;").values
node2 = pd.read_csv(node2_file, sep="\;").values
node3 = pd.read_csv(node3_file, sep="\;").values
# jammer = pd.read_csv(jammer_file, sep="\;").values

print(node2)
```

```python
node1_RSSI = (node1[(node1[:,0] == "RSSI")])[:,1]
node1_long = (node1[(node1[:,0] == "RSSI long")])[:,1]
node1_short = (node1[(node1[:,0] == "RSSI short")])[:,1]

node2_RSSI = (node2[(node2[:,0] == "RSSI")])[:,1]
node2_long = (node2[(node2[:,0] == "RSSI long")])[:,1]
node2_short = (node2[(node2[:,0] == "RSSI short")])[:,1]

node3_RSSI = (node3[(node3[:,0] == "RSSI")])[:,1]
node3_long = (node3[(node3[:,0] == "RSSI long")])[:,1]
node3_short = (node3[(node3[:,0] == "RSSI short")])[:,1]
```

```python
plt.plot(node1_RSSI, color='g', alpha=0.4)
plt.plot(node1_long, linewidth=3)
plt.plot(node1_short, linewidth=3)
plt.title("Node 1")
plt.xlabel("Measurement")
plt.ylabel("RSSI value (dBm)")
plt.legend(["Current RSSI value", "Moving average RSSI (60)", "Moving average RSSI (20)"], fontsize=14)
plt.savefig("node1.svg")
```

```python
plt.plot(node2_RSSI, color='g', alpha=0.4)
plt.plot(node2_long, linewidth=3)
plt.plot(node2_short, linewidth=3)
plt.title("Node 2")
plt.xlabel("Measurement")
plt.ylabel("RSSI value (dBm)")
plt.legend(["Current RSSI value", "Moving average RSSI (60)", "Moving average RSSI (20)"], fontsize=14)
plt.savefig("node2.svg")
```

```python
plt.plot(node3_RSSI, color='g', alpha=0.4)
plt.plot(node3_long, linewidth=3)
plt.plot(node3_short, linewidth=3)
plt.title("Node 3")
plt.xlabel("Measurement")
plt.ylabel("RSSI value (dBm)")
plt.legend(["Current RSSI value", "Moving average RSSI (60)", "Moving average RSSI (20)"], fontsize=14)
plt.savefig("node3.svg")
```
